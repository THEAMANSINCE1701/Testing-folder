```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sweetviz as sv
```

```python
df=pd.read_csv("matches.csv")
df1=pd.read_csv("deliveries.csv")
df
df1
```

```python
df1.head()
```

```python
df.head()
```

```python
ls=df["venue"].value_counts().sort_values(ascending=True)
ls
```

```python
ls.index # name of stadium
```

```python
ls.values #no. of matches 
```

```python
sns.set_style("darkgrid") #  give color to grid 
plt.figure(figsize=(15,8)) # give size of figure
temp= sns.barplot(ls.index,ls.values,alpha=0.8) #plotting barplot and defining x and y variables, alpha is a color parameter
plt.title("MOST PLAYED VENUE")
plt.xlabel("Name of stadium",fontsize=15)
plt.ylabel("count", fontsize=14)
temp.set_xticklabels(rotation=90,labels=ls.index,fontsize=10) #xtick label is used to give name of each x label.
plt.show()
```

```python
temp=pd.concat([df["umpire1"],df["umpire2"]]).value_counts().sort_values(ascending=True)
```

```python
temp.index
```

```python
temp.values
```

```python
sns.set_style("darkgrid") 
plt.figure(figsize=(25,8)) 
t= sns.barplot(temp.index,temp.values,alpha=0.9)
plt.title("favorite umpire")
plt.xlabel("Name of umpire",fontsize=15)
plt.ylabel("count", fontsize=15)
t.set_xticklabels(rotation=60,labels=temp.index,fontsize=10)
plt.show()
```

```python
ds=df.loc[(df["venue"]=="Feroz Shah Kotla")]
ws=ds[ds["win_by_runs"]>0]
slices=[len(ws),len(ds)-len(ws)]
lab=["batting 1","batting 2"]
plt.pie(slices,labels=lab,startangle=90,shadow=1,explode=(0,0.3),autopct="%0.3f%%",colors=["lightblue","pink"])
plt.title("feroz shah kotla")
plt.show()
```
