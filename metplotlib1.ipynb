```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#to give value to x-axis and y-axis
x=np.array([0,4,9,6])
y=np.array([0,250,350,450])
```

```python
#give values according to cooordinate 
plt.plot(x,y)
#function use to give show command 
plt.show()
```

```python
# o and * is a marker to plot dot chart
plt.plot(x,y,'*')
plt.show()
```

```python
# lifestyle use to change line style we can use it with marker too.
plt.plot (x,y,'o',linestyle ="dotted")
```

```python
plt.plot (x,y,'o',linestyle ="dashdot")
```

```python
plt.plot(x,y,color='r')
```

```python
plt.plot (x,y,'o',linestyle ="dotted",color='hotpink')
```

```python
plt.plot (x,y,linewidth=20)
```

```python
x=np.array([1,2,3,4,5])
y=np.array([50,70,100,200,800])
plt.plot(x,y,'o',linestyle ="dotted",color='hotpink')
plt.xlabel("average pulse")
plt.ylabel("calorie burned")
plt.title("stopwatch")
plt.grid()
plt.show()
```

```python
#scatter plot
x=np.array([1,2,3,4,5])
y=np.array([50,100,150,200,250])
colour=np.array([20,40,60,80,100])
plt.scatter(x,y,c=colour,cmap="viridis")
plt.colorbar()
plt.show()
```

```python
#bar pllot

x=np.array([30,87])
y=np.array(["apple","mangoes"])

plt.bar(y,x)
plt.xlabel("fruit name")
plt.ylabel("quantity")
plt.title("fruit vs quantity")
plt.show()
```

```python
x=np.random.randint(100,size=30)
print(x)
plt.hist(x,color='darkgreen')#histogram
plt.show()
```

```python
#piechart
x=np.array([23,55,66,56,])
l=["apple","mango","banana","orange"]
myex = [0.5,0,0,0]
plt.pie(x,labels=l, startangle=0,autopct='%0.0f%%',explode=myex,shadow=True)
plt.legend(title="fruits")
plt.title("pie chart")
plt.show()
```
